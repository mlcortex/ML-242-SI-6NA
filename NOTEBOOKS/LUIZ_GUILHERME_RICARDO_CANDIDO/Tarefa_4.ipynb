```python
import pandas as pd

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})

vinhos
```

```python
X = vinhos.iloc[:, 1:]
y = vinhos['classe']

print("Dimensão dos dados: ", vinhos.shape)
print(vinhos.head())
```

```python
from sklearn.model_selection import KFold

random_states = [42, 17, 24]
kf_config = {rs: KFold(n_splits=10, shuffle=True, random_state=rs) for rs in random_states}

for random_state, kf in kf_config.items():
    print(f"\nRandom State: {random_state}")
    for fold, (train_index, test_index) in enumerate(kf.split(X), 1):
        print(f"  Fold {fold}: {len(train_index)} treino, {len(test_index)} teste")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

ks = [3, 5]

# Lista para armazenar os resultados finais
resultados = []

# Itera sobre cada configuração de `random_state` e o objeto `kf_config`
for random_state, kf in kf_config.items():
    print(f"\nRandom State: {random_state}")

    for k in ks:
        acuracias = []

        # Executa o processo de cross-validation, dividindo o dataset em folds
        for train_index, test_index in kf.split(X):
            X_train, X_test = X.iloc[train_index], X.iloc[test_index]
            y_train, y_test = y.iloc[train_index], y.iloc[test_index]

            knn = KNeighborsClassifier(n_neighbors=k)
            knn.fit(X_train, y_train)

            y_pred = knn.predict(X_test)
            acuracia = accuracy_score(y_test, y_pred)
            acuracias.append(acuracia)

        # Calcula a média das acurácias de todos os folds
        media = sum(acuracias) / len(acuracias)
        desvio_padrao = pd.Series(acuracias).std()

        # Armazena os resultados em um dicionário e adiciona à lista `resultados`
        resultados.append({
            'random_state': random_state,
            'k': k,
            'media_acuracia': media,
            'desvio_padrao': desvio_padrao
        })

        # Exibe os resultados de média de acurácia e desvio padrão para o valor atual de `k`
        print(f"  k={k} -> Média da Acurácia: {media:.4f}, Desvio Padrão: {desvio_padrao:.4f}")
```

```python
import pandas as pd

# Converte a lista de resultados em um DataFrame para facilitar a análise
df_resultados = pd.DataFrame(resultados)

# Itera sobre os diferentes valores de `random_state` configurados
for random_state in random_states:
    print(f"\nOs resultados Random State {random_state}:")
    df_rs = df_resultados[df_resultados['random_state'] == random_state]
    print(df_rs)

    # Identifica a configuração com a maior média de acurácia para o random_state atual
    melhor_k = df_rs.loc[df_rs['media_acuracia'].idxmax()]
    print(f"A melhor configuração: k={melhor_k['k']}, Média da Acurácia: {melhor_k['media_acuracia']:.4f}")

# Identifica a melhor configuração geral, considerando todos os random_states
melhor_resultado = df_resultados.loc[df_resultados['media_acuracia'].idxmax()]
print("\nA Melhor configuração geral:")
print(melhor_resultado)
```
