```python
import pandas as pd

# URL do dataset Wine 
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})

# Exibir as primeiras linhas para confirmar o carregamento
print(vinhos.head())
```

```python
# 1. Análise da coluna classe
classe_counts = vinhos['classe'].value_counts()
print("Contagem das classes:")
print(classe_counts)
```

```python
# 2. Estatísticas descritivas para 'alcool'
alcool_stats = vinhos['alcool'].describe()
print("\nEstatísticas descritivas para 'alcool':")
print(alcool_stats)
```

```python
# 3. Estatísticas descritivas para 'acido_malico'
acido_malico_stats = vinhos['acido_malico'].describe()
print("\nEstatísticas descritivas para 'acido_malico':")
print(acido_malico_stats)
```

```python
# 4. Estatísticas descritivas para 'magnesio'
magnesio_stats = vinhos['magnesio'].describe()
print("\nEstatísticas descritivas para 'magnesio':")
print(magnesio_stats)
```

```python
# 5. Estatísticas descritivas para 'matiz'
matiz_stats = vinhos['matiz'].describe()
print("\nEstatísticas descritivas para 'matiz':")
print(matiz_stats)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Definir o tamanho da figura
plt.figure(figsize=(12, 8))

# Criar boxplots
columns_to_plot = ['magnesio', 'prolina', 'cinzas', 'alcool']
for i, col in enumerate(columns_to_plot):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(y=vinhos[col])
    plt.title(f'Boxplot de {col}')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

```python
vinhos.boxplot(column='magnesio', by='classe')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Criar o histograma
plt.figure(figsize=(8, 6))
sns.histplot(vinhos['fenois_totais'], bins=10)
plt.title('Histograma de Fenois Totais')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()
```

```python

import matplotlib.pyplot as plt
# Filtrar os dados para a classe 3
vinhos_classe_3 = vinhos[vinhos['classe'] == '3']

# Criar o histograma para a classe 3
plt.figure(figsize=(8, 6))
sns.histplot(vinhos_classe_3['fenois_totais'], bins=10)
plt.title('Histograma de Fenois Totais (Classe 3)')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()
```
