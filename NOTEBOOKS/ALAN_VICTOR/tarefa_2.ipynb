```python
import pandas as pd

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

```python
# Análise exploratória da coluna 'classe'
print("Contagem das diferentes classes:")
print(vinhos['classe'].value_counts())
print("\n")

# Análise exploratória das colunas numéricas: 'alcool', 'acido_malico', 'magnesio', 'matiz'
colunas_numericas = ['alcool', 'acido_malico', 'magnesio', 'matiz']
for coluna in colunas_numericas:
  print(f"Análise da coluna: {coluna}")
  print(vinhos[coluna].describe())
  print("\n")

# Exemplo de cálculo da correlação:
correlacao = vinhos[colunas_numericas].corr()
print("Matriz de correlação:")
print(correlacao)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Lista de colunas para criar boxplots
colunas_boxplot = ['magnesio', 'prolina', 'cinzas', 'alcool']

# Criar subplots para cada coluna
plt.figure(figsize=(15, 10))
for i, coluna in enumerate(colunas_boxplot):
  plt.subplot(2, 2, i + 1)
  sns.boxplot(x=vinhos[coluna])
  plt.title(f'Boxplot de {coluna}')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
# Lista de colunas para criar boxplots
colunas_boxplot = ['magnesio', 'fenois_totais', 'acido_malico', 'alcool']

# Criar subplots para cada coluna
plt.figure(figsize=(15, 10))
for i, coluna in enumerate(colunas_boxplot):
  plt.subplot(2, 2, i + 1)
  sns.boxplot(x='classe', y=coluna, data=vinhos)
  plt.title(f'Boxplot de {coluna} por Classe')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.hist(vinhos['fenois_totais'], bins=10)
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.title('Histograma da Distribuição de Fenois Totais')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.hist(vinhos[vinhos['classe'] == '3']['fenois_totais'], bins=10)
plt.xlabel('Fenois Totais (Classe 3)')
plt.ylabel('Frequência')
plt.title('Histograma da Distribuição de Fenois Totais (Classe 3)')
plt.show()
```
